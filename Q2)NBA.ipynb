NBA
The NBA is the highest paying professional sports league in the world," reported CNN in March 2016. The file nba_salaries.csv contains the salaries of all National Basketball Association players in 2015-2016

```python
'''Requirement-1 : Simple analysis'''
#1. Reading file as CSV

df = pd.read_csv("https://raw.githubusercontent.com/flaviovdf/evcomp2018/master/data/nba_salaries.csv")
df
```

```python
#2.Size
df.size
```

```python
#3. Sort on player

df.sort_values(by='PLAYER', ascending=True)
```

```python
#4. Sort on salary. Increasing and decreasing

df.sort_values(by='SALARY', ascending=True)
```

```python
df.sort_values(by='SALARY', ascending=False)
```

```python
#5. Relabel column salary to SALARY

df.rename(columns={'SALARY': 'SALARY'}, inplace=True)
```

```python
#6. Display 5 rows

df.head(5)
```

```python
#7. Select 3rd, 10th and 22nd rows.

df.iloc[[3, 10, 22]]
```

```python
#8. Top 5 highest paid

df.sort_values(by='SALARY', ascending=False).head(5)
```

```python
# 9.List those earned more than 10m USD.

df[df['SALARY'] > 10]
```

```python
#10.Earnings of Stephen Curry

df[df['PLAYER'] == 'Stephen Curry']
```

```python
#11. All Point Guards(PG) salary more than 15m USD.

df[(df['POSITION'] == 'PG') & (df['SALARY'] > 15)]
```

```python
#12. Team "Golden State Warriors"

df[df['TEAM'] == 'Golden State Warriors']
```

```python
''' Requirement-2 : Classifying by One Variable '''

"""13. How much money did each team pay for its players' salaries? (We have to group the
rows by TEAM  and then sum the salaries of the groups.)"""

df.groupby('TEAM').sum()['SALARY']
```

```python
"""14. How many NBA players were there in each of the five positions? (Classify by
POSITION , and count use grouping)"""

df.groupby('POSITION').count()['PLAYER']
```

```python
"""15. What was the average salary of the players at each of the five positions?(Group
by POSITION and take the mean of the salaries)."""

df.groupby('POSITION').mean()['SALARY']
```
